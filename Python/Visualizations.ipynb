# Visualizations

```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df= pd.read_csv("https://github.com/pablolopez2733/Predicting-Premier-League/raw/master/CleanData/TeamMetrics.csv")
df.head()
```

```python
df['LogPoints']=np.log(df['Points'])
df.head()
```

```python
all_seasons = df.groupby('team').mean().reset_index()
all_seasons.head()
```

```python
rcParams['figure.figsize'] = 8.4,6.46
ax = sns.regplot(x="AvgFifa", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('Ratings promedio de FIFA vs Log Puntos acumulados desde el ao 2014 hasta el 2020')
sns.set_style("whitegrid")


# figure size in inches
```

```python
rcParams['figure.figsize'] = 15,8
fig, axs = plt.subplots(ncols=2,nrows=2)
sns.scatterplot(x='AvgFifa', y='LogPoints', data=all_seasons,hue="team", ax=axs[0][0])
sns.scatterplot(x='AvgValue', y='LogPoints', data=all_seasons,hue="team", ax=axs[0][1],legend=False)
sns.scatterplot(x='Avg90', y='LogPoints', data=all_seasons,hue="team", ax=axs[1][0],legend=False)
sns.scatterplot(x='AvgRating', y='LogPoints', data=all_seasons,hue="team", ax=axs[1][1],legend=False)

plt.legend=axs[0][0].legend(loc='center right', bbox_to_anchor=(0, 0), borderaxespad=4.5)
#fig.legend(   # The labels for each line
       #    loc="center right",   # Position of legend
       #    borderaxespad=0.1,    # Small spacing around legend box
       #    title="Legend Title",  # Title for the legend
        #   ncol=2)

```
