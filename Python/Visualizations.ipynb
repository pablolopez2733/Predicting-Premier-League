# Visualizaciones

```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df= pd.read_csv("https://github.com/pablolopez2733/Predicting-Premier-League/raw/master/CleanData/TeamMetrics.csv")
df.head()
```

```python
#La variable respuesta es Log(points) para poder usar regresi√≥n lineal
df['LogPoints']=np.log(df['Points'])
df.head()
```

```python
#Tomemos los promedios de los datos durante las 6 temporadas para hacer un EDA
all_seasons = df.groupby('team').mean().reset_index()
all_seasons.head()
```

### EDA

```python
df_num = df.drop('team', axis=1)
df_num.head()
```

```python
def numeric_profiling(df, col, profiling_dict):
    """
    Profiling for numeric columns. 
    
    :param: column to analyze
    :return: dictionary
    """
    profiling = {}

    profiling.update({'Valor m√°ximo': df[col].max(),
                     'Valor M√≠nimo': df[col].min(),
                     'Media': df[col].mean(),
                     'Desv.Estandar': df[col].std(),
                     'Q1': df[col].quantile(.25),
                     'Mediana': df[col].median(),
                     'Q3': df[col].quantile(.75),
                     'Kurtosis': df[col].kurt(),
                     'Oblicuidad': df[col].skew(),
                     '# de valores unicos': df[col].nunique(),
                     '% de datos faltantes': sum(df[col].isna())/df[col].size*100,
                     'top1_repetidos': get_repeated_values(df, col, 1),
                     'top2_repetidos': get_repeated_values(df, col, 2),
                     'top3_repetidos': get_repeated_values(df, col, 3)})
    
    profiling_dict.update({col: profiling})
    
    return profiling_dict

def get_repeated_values(df, col, top):
    top_5 = df.groupby([col])[col]\
                    .count()\
                    .sort_values(ascending = False)\
                    .head()
    indexes_top_5 = top_5.index
    
    if ((top == 1) and (len(indexes_top_5) > 0)):
        return indexes_top_5[0]
    elif ((top == 2) and (len(indexes_top_5) > 1)):
        return indexes_top_5[1]
    elif ((top == 3) and (len(indexes_top_5) > 2)):
        return indexes_top_5[2]
    else: 
        return 'undefined'
    
variables = df_num.columns.values
profiling_dict = {}

numeric_profiling_output = [numeric_profiling(df, elem, profiling_dict)\
                            for elem in variables]
df_numeric_profiling = pd.DataFrame(numeric_profiling_output[0]).reset_index()
df_numeric_profiling.rename(columns={'index': 'metric'}, inplace=True)
```
