# Visualizations

```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df= pd.read_csv("https://github.com/pablolopez2733/Predicting-Premier-League/raw/master/CleanData/TeamMetrics.csv")
df.head()
```

```python
all_seasons = df.groupby('team').mean().reset_index()
all_seasons.head()
```

```python
rcParams['figure.figsize'] = 8.4,6.46
ax = sns.scatterplot(x="AvgFifa", y="Points", data=all_seasons,hue="team")
ax.set_title('Ratings promedio de FIFA vs Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
ax.legend(title='Equipos',bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,ncol=2)

# figure size in inches
```
