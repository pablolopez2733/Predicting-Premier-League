# Visualizaciones

```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
df= pd.read_csv("https://github.com/pablolopez2733/Predicting-Premier-League/raw/master/CleanData/TeamMetrics.csv")
df.head()
```

```python
#La variable respuesta es Log(points) para poder usar regresión lineal
df['LogPoints']=np.log(df['Points'])
df.head()
```

```python
#Tomemos los promedios de los datos durante las 6 temporadas para hacer un EDA
all_seasons = df.groupby('team').mean().reset_index()
all_seasons.head()
```

### Scatters

```python
#LogPoints vs AvgFifa

rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="AvgFifa", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('Ratings promedio de FIFA vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("AvgFifa.png")
```

```python
#LogPoints vs AvgValue
rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="AvgValue", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('Valor de mercado promedio del equipo vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("AvgValue.png")
```

```python
#LogPoints vs AvgRating
rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="AvgRating", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('Ratings Whoscoredpromedio vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("AvgRating.png")
```

```python
#LogPoints vs Avg90
rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="Avg90", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('Avg90 vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("Avg90.png")
```

```python
#LogPoints vs AvgGoals
rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="AvgGoal", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('AvgGoal vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("AvgGoal.png")
```

```python
#LogPoints vs PPG
rcParams['figure.figsize'] = 8.4,6.46
fig = plt.figure()
ax = sns.regplot(x="PPG", y="LogPoints",fit_reg=True, data=all_seasons)
ax.set_title('PPG vs Log Puntos acumulados desde el año 2014 hasta el 2020')
sns.set_style("whitegrid")
fig.savefig("PPG.png")
```

```python
#Summary
rcParams['figure.figsize'] = 15,8
fig, axs = plt.subplots(ncols=2,nrows=3)
sns.scatterplot(x='AvgFifa', y='LogPoints', data=all_seasons,hue="team", ax=axs[0][0],legend=False)
sns.scatterplot(x='AvgValue', y='LogPoints', data=all_seasons,hue="team", ax=axs[0][1],legend=False)
sns.scatterplot(x='Avg90', y='LogPoints', data=all_seasons,hue="team", ax=axs[1][0])
sns.scatterplot(x='AvgRating', y='LogPoints', data=all_seasons,hue="team", ax=axs[1][1],legend=False)
sns.scatterplot(x='AvgGoal', y='LogPoints', data=all_seasons,hue="team", ax=axs[2][0],legend=False)
sns.scatterplot(x='PPG', y='LogPoints', data=all_seasons,hue="team", ax=axs[2][1],legend=False)

plt.legend=axs[1][0].legend(loc='center right', bbox_to_anchor=(0, .6), borderaxespad=5.5,ncol=1)
fig.subplots_adjust(top=1.2, wspace=0.3)
fig.savefig("color_scatter.png")
```

### EDA

```python
df_num = df.drop('team', axis=1)
df_num.head()
```

```python
def numeric_profiling(df, col, profiling_dict):
    """
    Profiling for numeric columns. 
    
    :param: column to analyze
    :return: dictionary
    """
    profiling = {}

    profiling.update({'max': df[col].max(),
                     'min': df[col].min(),
                     'mean': df[col].mean(),
                     'stdv': df[col].std(),
                     '25%': df[col].quantile(.25),
                     'median': df[col].median(),
                     '75%': df[col].quantile(.75),
                     'kurtosis': df[col].kurt(),
                     'skewness': df[col].skew(),
                     'uniques': df[col].nunique(),
                     'prop_missings': sum(df[col].isna())/df[col].size*100,
                     'top1_repeated': get_repeated_values(df, col, 1),
                     'top2_repeated': get_repeated_values(df, col, 2),
                     'top3_repeated': get_repeated_values(df, col, 3)})
    
    profiling_dict.update({col: profiling})
    
    return profiling_dict

def get_repeated_values(df, col, top):
    top_5 = df.groupby([col])[col]\
                    .count()\
                    .sort_values(ascending = False)\
                    .head()
    indexes_top_5 = top_5.index
    
    if ((top == 1) and (len(indexes_top_5) > 0)):
        return indexes_top_5[0]
    elif ((top == 2) and (len(indexes_top_5) > 1)):
        return indexes_top_5[1]
    elif ((top == 3) and (len(indexes_top_5) > 2)):
        return indexes_top_5[2]
    else: 
        return 'undefined'
    
variables = df_num.columns.values
profiling_dict = {}

numeric_profiling_output = [numeric_profiling(df, elem, profiling_dict)\
                            for elem in variables]
df_numeric_profiling = pd.DataFrame(numeric_profiling_output[0]).reset_index()
df_numeric_profiling.rename(columns={'index': 'metric'}, inplace=True)
```
