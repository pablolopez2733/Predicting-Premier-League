## Create custom metrics based on goals allowed goals scored and minutes played

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
df17_18 = pd.read_csv("epldata_final.csv")
df17_18.head()
```

```python
fbref_19_20 = pd.read_csv("fbref.csv")
fbref_18_19 = pd.read_csv("fbref2.csv")
fbref_17_18 = pd.read_csv("fbref3.csv")
```

```python
metrics_17_18=fbref_17_18[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_17_18)
metrics_18_19=fbref_18_19[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_18_19)
metrics_19_20=fbref_19_20[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_19_20)
```

#### We'll just take each team´s 15 players with most minutes played

```python

pd.set_option("display.max_rows",None)

gk = metrics_17_18.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk2 = metrics_18_19.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk3 = metrics_19_20.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)

display(gk)  
```

```python
team_plus_minus_1718=gk.groupby('Squad')["+/-90"].mean()
team_plus_minus_1819=gk2.groupby('Squad')["+/-90"].mean()
team_plus_minus_1920=gk3.groupby('Squad')["+/-90"].mean()
display(team_plus_minus_1718)
display(team_plus_minus_1819)
display(team_plus_minus_1920)
```

```python
#Calcular cociente entre goles recibidos 
gk["GCoc"]=gk["onG"]/gk["onGA"]
gk2["GCoc"]=gk2["onG"]/gk2["onGA"]
gk3["GCoc"]=gk3["onG"]/gk3["onGA"]
display(gk)
```

```python
metrics1718=gk[['Player','Squad','+/-90','GCoc']].copy()
metrics1819=gk2[['Player','Squad','+/-90','GCoc']].copy()
metrics1920=gk3[['Player','Squad','+/-90','GCoc']].copy()
```

```python
team_gcoc_1718=gk.groupby('Squad')["GCoc"].mean()
team_gcoc_1819=gk2.groupby('Squad')["GCoc"].mean()
team_gcoc_1920=gk3.groupby('Squad')["GCoc"].mean()
display(team_gcoc_1718)
display(team_gcoc_1819)
display(team_gcoc_1920)
```

```python
team_metrics1718=pd.merge(team_gcoc_1718, team_plus_minus_1718, on='Squad')
team_metrics1819=pd.merge(team_gcoc_1819, team_plus_minus_1819, on='Squad')
team_metrics1920=pd.merge(team_gcoc_1920, team_plus_minus_1920, on='Squad')
```
