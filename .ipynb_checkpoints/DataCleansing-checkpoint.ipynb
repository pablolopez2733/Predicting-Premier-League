## Create custom metrics based on goals allowed goals scored and minutes played

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
df17_18 = pd.read_csv("Raw-Data/epldata_final.csv")
df17_18.head()
```

```python
fbref_19_20 = pd.read_csv("Raw-Data/fbref.csv")
fbref_18_19 = pd.read_csv("Raw-Data/fbref2.csv")
fbref_17_18 = pd.read_csv("Raw-Data/fbref3.csv")

fbref1415=pd.read_csv("Raw-Data/fbref1415.csv")
fbref1516=pd.read_csv("Raw-Data/fbref1516.csv")
fbref1617=pd.read_csv("Raw-Data/fbref1617.csv")
```

```python
metrics_17_18=fbref_17_18[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_17_18)
metrics_18_19=fbref_18_19[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_18_19)
metrics_19_20=fbref_19_20[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
display(metrics_19_20)

metrics_14_15=fbref1415[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
metrics_15_16=fbref1516[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
metrics_16_17=fbref1617[['Player', 'Squad', 'Min','PPM','onG','onGA','+/-','+/-90','On-Off']].copy()
```

#### We'll just take each team´s 15 players with most minutes played

```python

pd.set_option("display.max_rows",None)

gk = metrics_17_18.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk2 = metrics_18_19.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk3 = metrics_19_20.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)

gk4 = metrics_14_15.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk5 = metrics_15_16.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)
gk6 = metrics_16_17.groupby('Squad').apply(lambda grp: grp.nlargest(15,'Min')).reset_index(drop=True)

display(gk)  
```

```python
team_plus_minus_1718=gk.groupby('Squad')["+/-90"].mean()
team_plus_minus_1819=gk2.groupby('Squad')["+/-90"].mean()
team_plus_minus_1920=gk3.groupby('Squad')["+/-90"].mean()
team_plus_minus_1415=gk4.groupby('Squad')["+/-90"].mean()
team_plus_minus_1516=gk5.groupby('Squad')["+/-90"].mean()
team_plus_minus_1617=gk6.groupby('Squad')["+/-90"].mean()
display(team_plus_minus_1415)
display(team_plus_minus_1516)
display(team_plus_minus_1617)
display(team_plus_minus_1718)
display(team_plus_minus_1819)
display(team_plus_minus_1920)
```

```python
#Calcular cociente entre goles recibidos 
gk["GCoc"]=gk["onG"]/gk["onGA"]
gk2["GCoc"]=gk2["onG"]/gk2["onGA"]
gk3["GCoc"]=gk3["onG"]/gk3["onGA"]
gk4["GCoc"]=gk4["onG"]/gk4["onGA"]
gk5["GCoc"]=gk5["onG"]/gk5["onGA"]
gk6["GCoc"]=gk6["onG"]/gk6["onGA"]
display(gk6)
```

```python
metrics1415=gk4[['Min','Player','Squad','+/-90','GCoc']].copy()
metrics1516=gk5[['Min','Player','Squad','+/-90','GCoc']].copy()
metrics1617=gk6[['Min','Player','Squad','+/-90','GCoc']].copy()

metrics1718=gk[['Min','Player','Squad','+/-90','GCoc']].copy()
metrics1819=gk2[['Min','Player','Squad','+/-90','GCoc']].copy()
metrics1920=gk3[['Min','Player','Squad','+/-90','GCoc']].copy()
metrics1415
```

```python
team_gcoc_1415=gk4.groupby('Squad')["GCoc"].mean()
team_gcoc_1516=gk5.groupby('Squad')["GCoc"].mean()
team_gcoc_1617=gk6.groupby('Squad')["GCoc"].mean()
team_gcoc_1718=gk.groupby('Squad')["GCoc"].mean()
team_gcoc_1819=gk2.groupby('Squad')["GCoc"].mean()
team_gcoc_1920=gk3.groupby('Squad')["GCoc"].mean()
display(team_gcoc_1415)
display(team_gcoc_1516)
display(team_gcoc_1617)
display(team_gcoc_1718)
display(team_gcoc_1819)
display(team_gcoc_1920)
```

```python
team_metrics1415=pd.merge(team_gcoc_1415, team_plus_minus_1415, on='Squad')
team_metrics1516=pd.merge(team_gcoc_1516, team_plus_minus_1516, on='Squad')
team_metrics1617=pd.merge(team_gcoc_1617, team_plus_minus_1617, on='Squad')

team_metrics1718=pd.merge(team_gcoc_1718, team_plus_minus_1718, on='Squad')
team_metrics1819=pd.merge(team_gcoc_1819, team_plus_minus_1819, on='Squad')
team_metrics1920=pd.merge(team_gcoc_1920, team_plus_minus_1920, on='Squad')
```
